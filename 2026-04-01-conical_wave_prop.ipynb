```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as colors
```

```python
r = np.linspace(1e-3, 1, 1000)
z = np.linspace(0, 1, 1001)

rr, zz = np.meshgrid(r, z)
```

```python
kappa0 = 10
beta = 10

a_minus = 1
a_plus = 1
```

```python
prefix = 1/np.sqrt(2*np.pi*kappa0 * r)
E_in = a_minus * np.exp(1j * np.pi / 4) * np.exp(-1j * kappa0 * rr) * np.exp(1j * beta * zz) * prefix
E_out = a_plus * np.exp(-1j * np.pi / 4) * np.exp(1j * kappa0 * rr) * np.exp(1j * beta * zz) * prefix

omega_t = 0

E_tot = (E_in+E_out) * np.exp(-1j * omega_t)

I_tot = np.real(E_tot)**2
```

```python
fig, ax = plt.subplots()
extent = [min(z), max(z), min(r), max(r)]
im = ax.imshow(I_tot.T, extent=extent, origin="lower", vmax=0.1)
fig.colorbar(im)
ax.set_xlabel("z")
ax.set_ylabel("r")
```
